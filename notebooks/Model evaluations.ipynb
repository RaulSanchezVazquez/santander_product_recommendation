```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from src import config
```

```python
challengers = ['item_most_popular', 'user_most_popular', 'net2', 'net_multilabel']
ndcg_all, ndcg = {}, {}
for challenger_name in challengers:
    path = os.path.join(
        config.PRJ_DIR, 
        'data/processed/ranks/{0}/ndcg.csv'.format(challenger_name))
    
    ndcg_all[challenger_name] = pd.read_csv(path)['all_prod']
    ndcg[challenger_name] = pd.read_csv(path)['prods']

ndcg = pd.DataFrame(ndcg)
ndcg_all = pd.DataFrame(ndcg_all)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

ndcg.plot(
    marker='o',
    grid=True,
    ax=ax)

ax.set_title('Normalized Discounted Cumulative Gain\n(NDCG@10)')
ax.set_ylabel('NDCG')
ax.set_xlabel('Fold-id')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

ndcg_all.plot(
    marker='o',
    grid=True,
    ax=ax)

ax.set_title('Normalized Discounted Cumulative Gain\n(NDCG@10)')
ax.set_ylabel('NDCG')
ax.set_xlabel('Fold-id')
```
